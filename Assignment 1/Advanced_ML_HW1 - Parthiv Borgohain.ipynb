# Question 4: Linear Regression (10 pts)

1. What is the difference between R-square and adjusted R square and why is it desirable to use the adjusted value? (4 pts)

2. Overfitting usually happens in complex models. Linear Regression is a fairly simple model. Could overfitting happen in Linear Regression? If so, please explain the scenario in which it could happen and how we can tackle it. (6 pts)


## Answer

1. R-Square is given by: R^2=1-  RSS/TSS
   Adjusted R-Square is given by: R^2=1-  ((1 - R^2)(N - 1))/((N - p - 1))
   Here RSS = Sum of Square of Residuals
   TSS = Total Sum of Squares
   N = No. of data points
   P= No. of dependent variables

    Whenever new variables are added to a Regression Model, there is always an increase in R-Square value irrespective of whether the new variables are significant or not. Hence this makes R-Square a not-very-useful metric for checking whether a new variables is improving the model fit or not. 
    To take of care of this, the Adjusted R-Square Metric is used as it places a penalty on adding new unimportant variables (the term p in the denominator takes care of this). This makes Adjusted R-Square a better metric when looking at a model with multiple dependent variables.
    
    
    

2. Overfitting could happen in Linear Regression. For example, while sampling, if the training data set contains many data points that do not adhere to the actual relationship between variables, then the linear regression model would overfit. Also, a very large outlier value can cause overfitting. Outlier correction and Cross Validation can help us reduce overfitting.


Also, overfitting can happen if the no. of predictors is comparable to or larger than the no. of training data points. When this happens, small changes in the training set can lead to large variances between models. To take care of this, we could use regularization methods like Lasso or Ridge Regression as it would shrink the coefficients of unimportant variables and lead to reduction in variance.

```python
#Part 1
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import scale
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=50) #Splitting into train-test data
X_train=scale(X_train)
X_test=scale(X_test) #Scaled test data too
X_train[:5]
```

In Python, you can consider this range of values as follows: alpha = 10**numpy.linspace(6,-6,100) so that you can generate 100 uniform values between -6 to 6 as power series.

Fit the 2 regression models with scaled data and report the best chosen  based on cross validation as well as the corresponding scoring metric. The cross validation should happen on your training data using MSE as the scoring metric. (5 pts) 

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
from sklearn.linear_model import Lasso, Ridge, ElasticNet, LinearRegression
alpha = 10**np.linspace(6,-6,100)
lasso_model=Lasso()
grid = {'alpha':alpha}
gscv = GridSearchCV(lasso_model, param_grid=grid, scoring='neg_mean_squared_error', cv=5) #cross validation with k=5
results_lasso = gscv.fit(X_train, y_train)
print('MSE for Lasso: {:0.2f}'.format(-1*results_lasso.best_score_))
print('Best Alpha Value for Lasso: %s' % results_lasso.best_params_)
#####################################################################################

#Now running Ridge Regularization
ridge_model=Ridge()
gscv = GridSearchCV(ridge_model, param_grid=grid, scoring='neg_mean_squared_error', cv=5)
results_ridge = gscv.fit(X_train, y_train)
print('MSE for Ridge: {:0.2f}'.format(-1*results_ridge.best_score_))
print('Best Alpha Value for Ridge: %s' % results_ridge.best_params_)
```

**Part 4b**

**Ridge Regression**

Pros: Ridge Regression prevents overfitting in Linear Regression by shrinking the coefficients towards 0. If the target variable is a function of many dependent variables, all with coefficients of similar size, then Ridge Regression performs better.

Cons: Ridge Regression does not perform feature selection as it does not make any of the coefficients 0. Although it will shrink the coefficients, the coefficients never truly become 0. So, in a setting where a relatively small number of coefficients have substantial effects, Ridge Regression does not perform as good as Lasso Regression


**Lasso Regression**

Pros: Lasso Regression, like Ridge Regression also prevents overfitting in Linear Regression by shrinking the coefficients towards 0. However, it can do feature selection by setting the coefficients of unimportant variables to 0. It does better than Ridge Regression in settings where a relatively small number of coefficients have substantial impact on the target variable.

Cons: The features selected will be highly biased. If n << p, Lasso will select at max n features even if all the features are important. If there are a number of highly correlated variables, Lasso will select only of them arbitrarily. Also, in general, perforamce of Lasso Regression is worse than Ridge Regression.

**Elastic Net**

Pros: It combines the advantages of Ridge and Lasso Regularization. Unlike Lasso Regression, it can select more than n predictors when n << p

Cons: It is computationally more expensive than both Lasso and Ridge.
