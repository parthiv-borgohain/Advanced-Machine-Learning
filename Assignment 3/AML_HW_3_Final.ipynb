# Question 2 : t-SNE (7 pts)

Read this [article](https://erdem.pl/2020/04/t-sne-clearly-explained) on t-SNE, a dimensionality reduction technique for visualization and explain it in your own words in one or two paragraphs.

#### Answer:

t-SNE is a **non-linear dimensionality reduction** technique. This algorithm allows us to separate data that cannot be separated by any straight line. In the first stage, the euclidean distances of every pair of points is calculated. Then, these distances are transformed into conditional probabilities that represent the similarity between every two points. The conditional probability ($p_{j|i}$) of point $x_{i}$ being next to point $x_{j}$ is calculated by assuming a Gaussian centered at $x_{i}$ with a standard deviation of $\sigma_{i}$. $\sigma_{i}$ is computed on the basis of the **Perplexity** parameter which can be interpreted as the no. of effective neighbours for each point. To calculate the conditional probability, we also divide the gaussian probability by the sum of all the probabilities of all the other points placed at the Gaussian centered at $x_{i}$ so that we can deal with clusters of different densities. 

From the conditional distributions created, we calculate the joint probability distribution for $p_{ij}$ as the mean of $p_{i|j}$ and $p_{j|i}$.

In the next stage, we create a dataset of points in a low-dimensional space and calculate a joint probability distribution for them as well. To do this, we build a random dataset of points with the same number of points as we had in the original dataset, and K features, where K is our target dimension. For this set of points, we will calculate their joint probability distribution by using the t-distribution and not the Gaussian distribution.

In the final stage, the **Kullback-Leiber divergence** is utilized to make the joint probability distribution of the data points in the low dimension as similar as possible to the one from the original dataset. If this transformation succeeds, we get a good dimension reduction of the original input data.


# Question 3: PCA (8 pts)

Explain the principle of Principal Component Analysis algorithm and in what sense is the selection of the eigenvectors to represent the data is an optimal choice.  How do you reconstruct (a noisy version of) the original data from the eigenvectors and the loadings?



#### Answer:


PCA aims to create a lower-dimensional representation that best represents the variance of the whole dataset. It uses Singular Value Decomposition to project data points to a lower dimension by trying to find a linear combination of inputs (called Principal Components) such that they capture most of the variance (information) present in the dataset. It is a dimensionality reduction technique. 

Eigenvalue decomposition of the Covariance Matrix gives both eigen values and eigenvectors. Hence, finding the eigenvalues and eigenvectors of the covariance matrix is the same as fitting the straight, principal component lines to data. As the eigenvectors trace the principal lines of force, and the axes of greatest variance and covariance illustrate where the data is most susceptible to change. By ranking the eigenvectors in order of their eigenvalues, highest to lowest, we get the principal components in order of significance.

In order to reconstruct (a noisy version of) the original data from the eigenvectors and the scores, we can use:

$$PCA\,Reconstruction= PC\,Scores‚ãÖEigenvectors^‚ä§ + Mean$$

Where $PC\,Scores =  XV$

and **X** is the centered Data Matrix (obtained after subtracting the mean vector **$\mu$** from each row) and **V** is the matrix of some k eigenvectors (PCs) that we want to use (usually the k eigenvectors with the k largest eigenvalues). The reconstruction obtained is a noisy reconstruction of the original input data.

e) i)**(2pts)** Calculate 95% confidence intervals for SalePrice with a sample size of 100. 


````python
import scipy
np.random.seed(10)
sample_size = 100
sample = np.random.choice(a= data_dm['SalePrice'],
                          size = sample_size)
sample_mean = sample.mean()

#Get the critical Z value

z_critical = scipy.stats.norm.ppf(1-.05/2) #two tailed

#Get population standard deviation
population_mean = data_dm['SalePrice'].mean()
pop_stdev = data_dm['SalePrice'].std()

#margin of error
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size)) 

#confidence interval
confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

#Print confidence interval and true mean value
print("95% Confidence Interval for Sample Mean is: ",confidence_interval)
print("True Population Mean is: ",population_mean)
````

ii)**(3pts)** Calculate 95 % confidence intervals for 100 different trials with a sample size of 500. Plot the confidence intervals and interpret how it captures the population mean.

````python
np.random.seed(12)

#define the sample size
sample_size = 500

intervals = []
sample_means = []

for sample in range(100):
    sample = np.random.choice(a= data_dm['SalePrice'], size = sample_size)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)
    #compute z critical value
    z_critical = scipy.stats.norm.ppf(1-.05/2)
    #compute population std dev        
    pop_stdev = data_dm["SalePrice"].std()
    
    margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))
    confidence_interval = (sample_mean - margin_of_error,
                           sample_mean + margin_of_error)  
    
    intervals.append(confidence_interval)
    

plt.figure(figsize=(13, 9))

plt.errorbar(x=np.arange(0.1, 100, 1), 
             y=sample_means, 
             yerr=[(top-bot)/2 for top,bot in intervals],
             fmt='o')

plt.hlines(xmin=0, xmax=100,
           y=data['SalePrice'].mean(), 
           linewidth=2.0,
           color="red")
plt.title('Confidence Intervals for 100 Trials', fontsize = 20)
plt.show()
````

### Solution:

The 95% confidence intervals tell us that if  we repeat the same experiment many times and draw many sample means (along with their confidence intervals), the confidence intervals created will contain the population mean about 95% of the time.In this plot, we are plotting 95% confidence intervals for 100 trials with sample size 500. From the plot, we can see that the population mean exists within approx 95% of these intervals. 

Thus, we can clearly observe that the 95% confidence intervals will capture the population mean about 95% of the time.
